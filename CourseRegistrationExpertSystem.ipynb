```
import sys
from experta import *

Year = ' ' 
TotalHours = 0

def Search(arr, number): 
  for i in range (len(arr)):
    if(int(arr[i].number) == int(number)): 
      return i

class Hourse(Fact):
  pass 

class Year(Fact):
  Year = Field(str, mandatory = True)

class Course(Fact):
  def __init__(self, name, code, number, credit, pre):
    self.name = name
    self.code = code
    self.number = number 
    self.credit = credit
    self.pre = pre

class AddedCourse(Fact):
  def __init__(self, name, code, number, credit, pre):
    self.name = name
    self.code = code
    self.number = number 
    self.credit = credit
    self.pre = pre

class Faculty(Fact):
  pass

class Major(Fact):
  pass

class PreCourse(Fact):
  pass

AddedCourses= []

ThirdYearCourses= [Course("Applied Math", "CPCS", 212, 6, None),\
          Course("Digital Logic Design", "CPCS", 211, 5, None),\
          Course("Data Base", "CPCS", 241, 5, None),\
          Course("Computer Organization and Acrchtecture", "CPCS", 214, 5, "CPCS 211"),\
          Course("Analysis and Design of Algorithm", "CPCS", 223, 5, None),\
          Course("Lab Science", " ", 200 , 6, None),\
          Course("Algorithms and Data Structure", "CPCS", 324, 5, "CPCS 223"),\
          Course("Software Engineering", "CPCS", 351, 5, None),\
          Course("Applied Probability & Random Process", "STAT", 352, 5, None)]

PreThirdYearCourses=[None,None,None, ThirdYearCourses[1],None,None,ThirdYearCourses[4],None,None]


FourthYearCourses= [Course("Programming languages", "CPCS", 301, 5, None),\
           Course("Artifcial Inteligent", "CPCS", 331, 5, "CPCS 223"),\
           Course("Intoduction to Software Project Management", "CPIS", 334, 3, None),\
           Course("Operaring Systems", "CPCS", 361, 5, "CPCS 214"),\
           Course("Computer Networks", "CPCS", 371, 5, None),\
           Course("Compiler Construction", "CPCS", 302, 5, None),\
           Course("Human-Computer Interaction", "CPCS", 381, 3, "CPCS 301"),\
           Course("workplace Skills", "CPIS", 393, 4, None)]

PreFourthYearCourses=[None, ThirdYearCourses[4], None, ThirdYearCourses[3],None,None,FourthYearCourses[0],None]


FifthYearCourses=[Course("Senior Project 1", "CPCS", 498, 3, None),\
         Course("Computer Graphics", "CPCS", 391, 5, None),\
         Course("Senior Project 1", "CPIS", 499, 3, "CPCS 498"),\
         Course("Profrssional Computing Issues", "CPIS", 428, 3, None)]
         
PreFifthYearCourses=[None, None, FifthYearCourses[0],None]


class CourseKB(KnowledgeEngine):

  @Rule(Faculty(FCIT = 'n'))
  def FacultyName(self):
    print("We're sorry we can't help you, the system is only for FCIT students!")
    sys.exit()


  @Rule(Faculty(FCIT = 'y'), Major(CS='n'))
  def MajorName(self):
    print("We're sorry we can't help you, the system is only for CS major students!")
    sys.exit()


  @Rule(Faculty(FCIT = 'y'),Major(CS='y'),Year(Year= 't')) 
  def ThirdYear(self):
    print("\n________________________________________________________________________________________________________________\n\n\%-50s%-15s%-15s%-15s%-15s\n\________________________________________________________________________________________________________________\n"\
          % ("Name", "Code", "Number", "Credit", "Pre"))
                  
    for i in range (0, len(ThirdYearCourses)):
      print(" %-50s%-20s%-15s%-15s%-15s" % (ThirdYearCourses[i].name, ThirdYearCourses[i].code, ThirdYearCourses[i].number , ThirdYearCourses[i].credit,  ThirdYearCourses[i].pre))              
      print("\n")


  @Rule(Faculty(FCIT = 'y'), Major(CS='y'), Year(Year= 'f'))
  def FourthY(self):
    print("\n________________________________________________________________________________________________________________\n\n\%-50s%-15s%-15s%-15s%-15s\n\________________________________________________________________________________________________________________\n"\
          % ("Name", "Code", "Number", "Credit", "Pre"))
                  
    for i in range (0, len(FourthYearCourses)):
       print(" %-50s%-20s%-15s%-15s%-15s" % (FourthYearCourses[i].name, FourthYearCourses[i].code, FourthYearCourses[i].number, FourthYearCourses[i].credit, FourthYearCourses[i].pre))              
       print("\n")


  @Rule(Faculty(FCIT = 'y'), Major(CS='y'), Year(Year= 'v'))
  def FifthY(self):
    print("\n________________________________________________________________________________________________________________\n\n\%-50s%-15s%-15s%-15s%-15s\n\________________________________________________________________________________________________________________\n"\
          % ("Name", "Code", "Number", "Credit", "Pre"))
                  
    for i in range (0, len(FifthYearCourses)):
       print(" %-50s%-20s%-15s%-15s%-15s" % (FifthYearCourses[i].name, FifthYearCourses[i].code, FifthYearCourses[i].number, FifthYearCourses[i].credit, FifthYearCourses[i].pre))              
       print("\n")


  @Rule(Year(Year = 't'), Hourse(max = 18))
  def MaxReach3(self):
    print(" Can’t add the course, you have reached the maximum hours in the semester!!!")


  @Rule(Year(Year = 'f'), Hourse(max = 18))
  def MaxReach4(self):
    print("Can’t add the course, you have reached the maximum hours in the semester!!!")


  @Rule(Year(Year = 'v'), Hourse(max = 20))
  def MaxReach5(self):
    print("Can’t add the course, you have reached the maximum hours in the semester!!!")

  @Rule(PreCourse(precourse = 'n'))
  def precourseNo(self):
    print("Can’t add the course, you need to study the pre first!!!")


engine = CourseKB()
engine.reset()


while (True):

  IsFCIT = input("Are you a student at FCIT? (Y/N): ").lower()
  if IsFCIT == 'y':
    IsCS = input("Are you specialized in Computer Science? (Y/N): ").lower()
    if IsCS == 'y':
      TheYear = input("What Year are currently studying in? \n\
                      ( 3rd Year Enter 3\n\
                        4th Year Enter 4\n\
                        5th Year Enter 5):  ").lower()
      
      while (TheYear != '3' and TheYear != '4' and TheYear !='5') :
        print("Please Enter a correct choice ")
        TheYear = input("What Year are currently studying in? \n\
                      ( 3rd Year Enter 3\n\
                        4th Year Enter 4\n\
                        5th Year Enter 5):  ").lower()
          
      engine.declare(Faculty(FCIT = IsFCIT), Major(CS = IsCS), Year(Year = TheYear))
      engine.run()
      break

    elif IsCS == 'n':
      engine.declare(Faculty(FCIT = IsFCIT), Major(CS= IsCS)) 
      engine.run()
      break 

    else:
        print("Please Enter a correct choice ")
        continue

  elif IsFCIT == 'n' :
    engine.declare(Faculty(FCIT = IsFCIT))
    engine.run()
    break

  else:
    print("Please enter a valid answer ")
    continue


while True:
  if (TotalHours >= 18):
    engine.declare(Year(Year = TheYear), Hourse(max = 18))
    engine.run()

  else:
    ChooseCourse = input("Enter the number of the course that you want to add: ")
    
    if (TheYear == '3' ):
      i = Search(ThirdYearCourses,ChooseCourse)

      if((TotalHours+ ThirdYearCourses[i].credit) > 18):
        engine.declare(Year(Year = TheYear), Hourse(max = 18))
        engine.run()

      else:  
        if PreThirdYearCourses[i] != None :
          print("The course you choose has Pre Courses, ", PreThirdYearCourses[i].code, PreThirdYearCourses[i].number)
          TookPre = input("Did you take this course before? (y/n): ").lower()


          if TookPre == 'n':
            engine.declare(PreCourse(precourse = TookPre))
            engine.run()
          else:
            print("The course have been added successfully") 
            TotalHours+= ThirdYearCourses[i].credit
            
        else:
          print("The course have been added successfully") 
          TotalHours+= ThirdYearCourses[i].credit

    elif (TheYear == '4'):
      i = Search(FourthYearCourses,ChooseCourse)

      if((TotalHours+ FourthYearCourses[i].credit) > 18):
        engine.declare(Year(Year = TheYear), Hourse(max = 18))
        engine.run()
      else:
        if PreFourthYearCourses[i] != None :

          print("The course you choose has Pre Courses, ", PreFourthYearCourses[i].code , PreFourthYearCourses[i].number)
          TookPre= input("Did you take this course before? (y/n): ").lower()
          if TookPre == 'n':
            engine.declare(PreCourse(precourse = TookPre))
            engine.run()
          else:
            print("The course have been added successfully") 
            TotalHours+= FourthYearCourses[i].credit

        else:
          print("The course have been added successfully")   
          TotalHours+= FourthYearCourses[i].credit
    else:
      i = Search(FifthYearCourses,ChooseCourse)
      
      if((TotalHours+ FifthYearCourses[i].credit) > 20):
        engine.declare(Year(Year = TheYear), Hourse(max = 20))
        engine.run()
      else:
        if PreFifthYearCourses[i] != None :
          print("The course you choose has Pre Courses, ", PreFifthYearCourses[i].code, PreFifthYearCourses[i].number)
          TookPre= input("Did you take this course before? (y/n): ").lower()
          
          if TookPre == 'n':
            engine.declare(PreCourse(precourse = TookPre))
            engine.run()
          else:
            print("The course have been added successfully") 
            TotalHours += FifthYearCourses[i].credit

        else:
          print("The course have been added successfully")
          TotalHours+= FifthYearCourses[i].credit

    if (TotalHours< 18):
      addAnother = input("Do you want to add another Course? (y/n): ").lower()
      while (addAnother != "y" and addAnother != "n"):
        print("Please enter a valid answer")
        addAnother = input("Do you want to add another Course? (y/n): ").lower() 
      if addAnother == "n":
        print("Thank you for using our system")
        break
    else:
      print("You have reached the maximum hours in the semester\n\Thank you for using our system")
      break


engine = CourseKB()
engine.run()
```
